### Identify and analyze patterns in the data.

```python
df.corr(numeric_only=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.heatmap(df.corr(numeric_only=True) , annot=True)
plt.show()
```

### Visualize the distribution of key variables (e.g., Traffic_Index, Date).

```python
plt.hist(df['JamsLength']);
plt.xlabel('JamsLength')
plt.ylabel('Frequency')
plt.title('Histogram of JamsLength')
plt.show()
```

```python
plt.hist(df['TravelTimeLive']);
plt.xlabel('TravelTimeLive')
plt.ylabel('Frequency')
plt.title('Histogram of TravelTimeLive')
plt.show()
```

```python
plt.hist((df['JamsCount']));
plt.
```

### Explore relationships between variables (e.g., Traffic_Index vs. Weather_Condition).


```python
sns.scatterplot(x='JamsDelay', y='JamsLength', data=df);
plt.xlabel('Jams Delay')
plt.ylabel('Jams Length')
plt.title('Relationship between JamsDelay and JamsLength')
plt.show()
```

```python
sns.scatterplot(x='JamsCount', y='JamsDelay', data=df);
plt.xlabel('Jams Count')
plt.ylabel('Jams Delay')
plt.title('Relationship between Jams Count and Jams Delay')
plt.show()
```
